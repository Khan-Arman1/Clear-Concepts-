```python
# what are nested functions
# The fucntions that are enclosed within a function are Nested funciton
# Example: 
def enclosed_function():

    # scope of nested start
    def nested_funtion():
        print("Call Nested function")
    # scope of nested end

    print("Call Enclosed function")
```

**NonLocal Variable**

```python
# a variable that is define under enclosed function 
def enclosed_function():

    var = "Enclosed function" 
    print("before changing the value" + var)
    # scope of nested start
    def nested_funtion():
        var = "Nested function"
        print("Nested function Try to change the variable" + var)
    # scope of nested end
    print("after changing the value" + var)

    return nested_funtion()
```

```python
enclosed_function()
```

**Here nonlocal keyword come into existance**

```python
# the nonlocal keyword allows a nested function to access the original variable and maniputate the original value
def enclosed_function():

    var = "Enclosed function"
    print("before changing the value\t" + var)
    # scope of nested start
    def nested_funtion():
        nonlocal var
        var = "Nested function"
        print("Nested function Try to change the variable\t" + var)
    # scope of nested end

    nested_funtion()
    print("after changing the value\t" + var)
```

```python
enclosed_function()
# here the value is changed 
```

# Clouser Function

* before understanding a clouser first define the nested function that reads 1 or more nonlocal variables and return the nested function

```python
def forest(): # enclosed()
    animals = 200
    plants = 2000

    def population(): # nested()
        return animals
    return population

amazon = forest() # creating a object of fucntion not of the class(OOP), this is called Decorator
print(amazon) # object of forest()
print(amazon())
```

```python
# del forest
# forest()

amazon()
```

* in the above code the population() is a **clouser()** 
* lets delete the enclosed() and see what happens

```python
def forest(): # enclosed function
    animals = 200

    def population(): # nested function
        return animals
    return population

amazon = forest() # creating a object of fucntion not the class this is called Decorator
amazon
# delete the forest
# del forest() # gives error of can't delete funciton call : - function call is forest()
# del forest # this is how we delete a fucntion
```

**how this happend**
* bcz python create a enviornment for the nested() by connecting to the nonlocal variable
* This method of remembering the values of nonlocal variables even if the variables are no longer in scope or the function is no longer in the current namespace is called Closure.

**Uses of Clouser()**

* The most common use case is in python decorator.
* The core principal of decorator is based on the clouser.

```python
def math(func):

   def div(numerator, denominator):
       if denominator == 0:
           return "Zero Division Error"
       return func(numerator, denominator)
   return div

@math
def division(a, b):
   return a/b

print(division(4, 2))
print(division(4, 0))
```
