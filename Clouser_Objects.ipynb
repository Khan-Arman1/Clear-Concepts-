# What is Clouser In Python

 * A closure is a function object that “remembers” the values of free variables from its enclosing scope even after that scope has finished execution.

 **Before understanding what is a clouser lets have a look on nested function**

 * **Nested Function** - Function inside a function or enclosed function.

 * **Enclosed Function** - The function have nested functions.

```python
def enclosed_function():
    # scope of nested start
    def nested_funtion():
        print("Call Nested function")
    # scope of nested end
    print("Call Enclosed function")
```

```python
# on calling the enclosed function
enclosed_function() # we can ass that function 
nested_function() 
```

#### Why got and error?
 
 * The scope of the 'nested_function()' started inside enclosed_function() and ended inside enclosed_function(). 
 * That function is inside and enclosed scope.
 * We are accessing the function in global scope.

##### Understand what are nonlocal variables 
 
 **NonLocal Variables** are defined inside an enclosed functions but outside of the nested function.

```python
def enclosed_function():

    var = "Enclosed function, Nonlocal Variable" 
    print("before changing the value" + var)

    # scope of nested start
    def nested_funtion():
        var = "Nested function, Local variable"
        print("Nested function Try to change the variable" + var)
    # scope of nested end
    print("after changing the value" + var)

    return nested_funtion()


enclosed_function()
```

#### What happened here?  
 * Python follows LEGB rule for variable scopes - 

        L - Local

        E - Enclosed 
        
        G - Global

        B - Built-in

 * Python first check for the variable in local scope, then search inside the enclosed scope, then in Global scope and in last searches in Built-in scope. 
 * If no variable found it raise NameError. 

```python
print(print) # Built-in scope
print(var) # without scope
```

The nonlocal keyword allows a nested function to access the original variable and maniputate the original value.

```python
def enclosed_function():

    var = "Enclosed function"
    print("before changing the value\t" + var)
    # scope of nested start
    def nested_funtion():
        nonlocal var
        var = "Nested function"
        print("Nested function Try to change the variable\t" + var)
    # scope of nested end

    nested_funtion()
    print("after changing the value\t" + var)

enclosed_function()
# here the value is changed 
```

# Clouser Function

* Before understanding a clouser first define the nested function that reads 1 or more nonlocal variables and return the nested function


```python
def forest(): # enclosed()
    animals = 200
    plants = 2000

    def population(): # nested()
        # Here the animal is in the nonlocal scope and we haven't define another same name variable, so it is accessible
        return animals
    return population

amazon = forest() # creating a object of fucntion not of the class(OOP), this is called Decorator
print(amazon) # object of forest()
print(amazon())
```

##### How it works?
 
 forest() ---> amazon ---> amazon(object) + () ---(function call)--> animals
 
 forest() ---> population ---> population(object) + () ---(function call)--> animals

**Uses of Clouser()**

* The most common use case is in python decorator.
* The core principal of decorator is based on the clouser.

```python
def math(func):

   def div(numerator, denominator):
       if denominator == 0:
           return "Zero Division Error"
       return func(numerator, denominator)
   return div

@math
def division(a, b):
   return a/b

print(division(4, 2))
print(division(4, 0))
```

```python
def world(): # enclosed function
    count = 0 # free variable

    def population(): # nested function
        nonlocal count # nonlocal ke bina modify nahi kar sakte
        count += 1 # free variable
        return count # free variable return
    
    return population

w = world() # execution ke baad destroy hoga but free variable and population() store ho jayenge
print(w())
print(w())
print(w())
print(w())
```
