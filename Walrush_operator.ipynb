```python
# helps to create a variable directly in the condition
# For example
while True:
    # ask user to enter a number
    num = input("Number :")
    if not num.isdigit():
        print("Enter a Valid Number.")
    else:
        break
    
```

```python
# with walrus operator you can convert that code in just few lines
while True:
    if not (num := input("Number: ").isdigit()):
        print("Enter a Valid Number.")
    else:
        break
# This is the key difference  
```

### **Convert string -> list and list -> string**

```python
st = "Arman Khan"
print(st)
print(type(st))
lst = list(st)
print(lst)
print(type(lst))
st = str(lst)
print(st)
print(type(st))
print("12".join(st))
print("12".join(lst))
```

### **Convert String -> tuple and tuple -> string**

```python
st = "Arman Khan"
print(st)
print(type(st))
tpl = tuple(st)
print(tpl)
print(type(tpl))
st = str(tpl)
print(st)
print(type(st))
print("12".join(st))
print("12".join(tpl))
```

### **Convert string -> set and set -> string**

```python
st = "Arman Khan"
print(st)
print(type(st))
st = set(st)
print(st)
print(type(st))
st = str(st)
print(st)
print(type(st))
print("12".join(st))
print("12".join(st))
```
