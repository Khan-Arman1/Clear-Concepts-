# What are Constructors in Python?

 * In python, Constructors are __ init __() methods.
 * But generally refer to the entire process how objects are create and initiallize.
 * **__ new __() Method -** 
     * Object creator.
     * Python automatically executes a method called '__ new __()' that creates the object.
     * '__ new __()' allocate memeory to the object.
     * And that object have its unique memory address.
     * Python automatically store reference of class in '__ new __()' method 'cls' argument.
 * **__ init __() Method -**
     * After executing '__ new __()' python instantly execute ' __ init __()'.
     * ' __init __()' initialize the object and provide state to objects.
     * Python then store reference of the created object as a parameter in 'self' argument.
* Methods and attributes of class provide behaviour to all the different objects.

* **In Python objects are actually created in Heap.**


# How Python store reference of objects?

 * Python does not expose memory location but provide reference to 'self/cls' or a variable.
     
     * Example - Suppose you ordered a pizza and you that delivery partner ask you for your exact location but you guide the delivery partner to reach a landmark and you yourself collect pizza from him. Here you did not tell him your exact address but you referenced the address.
     
 * Python actually use internally object reference system to access the data at that memory location.
 * Python creates a handle to reach that memory location and does not give the raw address.
 * In python classes, 'self' and 'cls' receive the reference(means handle) of the object and class.
     * That reference is internally connected to the memeory address.
 * *handle* - way to reach out

# How Python reference is different from C++ memory allocation?
 
 * As we know python uses a internally object reference system, a handle to reach that data.
 * But, in C++ we can access that memory location directly by use '&' operator.

    Example - int x = 10;   
              
              cout << &x;   

              Actual memory address

 * In Python, 

    Example - a = [1,2,3]
              
              b = a

              Here a and b reference to the same memeory location.

```python
a = [1,2,3]
b = [1,2,3]
c = a

# 'is' Identity operator - same as '==' but also checks reference (or say memeory location)
print(a is c)
print(a is b)
print(a == b)

print(id(a))
print(id(b))
print(id(c))
```

# How python internally store variables?

#### 1. Python Variable = Refrence Holder
 * In python variables do not store the raw address.
 * Variables store the pointer/reference.
 * When you create a variable - 
    * __ new __() - create object, allocate memeory (on RAM).

#### 2. Reference Count 
 * Reference Count means how many variables pointing/referencing to the same object.
 * Ever object have - 
    1. type - data type.
    2. value/state - actaul data.
    3. reference count - how many variables pointing the same object.
 * Variables internally store the pointer/refernce that refer to the object location.

```python
a = [1, 2, 3]
b = a   # b now references same object
```

* a and b both shares same reference.
* Object reference count increments internally.
* If we delete all pointers/references of an object, the object will also deleted from the memory.
* If a single reference is present the object will remain still in the memory

#### 3. Memeory Address Vs Reference

```python
print(id(a))
print(id(b))
print(id([1, 2, 3]))
```
