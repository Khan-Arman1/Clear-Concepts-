# AND Operator Confusion Statement -

### Suppose we have 2 variables


```python
a = 1
b = True
# lets print the and operation 
print(a and b)
print(b and a)
```

Remember this is not about numbers 

This is what about Python evaluates 'and' operator.

**As we have learned that and operator return true if both the statements are true.**

 * Here Both the statements are true because a is not null, non 0, non empty string and b is true.
 * But in here in (a and b) return true and in (b and a) returns 1 but Why? AND must return true or 1 in both the conditions.
 * **Answer-** Yes, it must return true or 1 in both the condition *But* AND takes one true and one false. 
    * In python, and does not return true or false    
    * It actually returns one of the operand.    
    * But When we see, 

        * a and b → since a truthy → return b
        * b and a → since b truthy → return a


#### **Question** - Both the variables (operands) are non zero, empty and false. But why does it return 1 and true in (a and b) and (b and a). Why?

#### **Answer-** Python checks truthiness but returns the original operand, not a boolean value.

 * In here a is truthy part -> b (to evaluate) 
 * In here b is truthy part -> a (to evaluate)
 * Here whether the value is true or false it evaluetes the data, If a is falsy part then it return output as a because its falsy part (false output)
 
 * When a is truthy return b to evaluate if b is falsy return b (false output)
 * When a and b both are truthy part then return the last evaluated truthy part as b


```python
# lets check in here
print("1" and 1 and 0) # return 0 
print("1" and 0 and True) # return 0 
print(" " and "Arman" and 534) # return 534 because " " space, "" empty
print("" and "Arman" and 534) # return false/nothing(space)
```

#### Moral - If any operand in 'and' expression is falsy, Python returns the first falsy operand. If not return the last value/operand.

 * Logical operators in Python return operands, not necessarily True or False.
 * It checks truthiness but returns the original operand.
 * It simply return one operand.

# Short-Circuit Evaluation

 * Works on 'and' and 'or' operator.
 * In Python, 'and' and 'or' operator does not evaluate the expression completely.
 * They stop early when they encounter the decide result.
 * This is called Short-Circuit Evaluation.

```python
# For example 
print("" and "4" and " ") 
```

## 'AND' Operator Short-Circuit

 * In and operator, 'and' evaluates the expression when it encounters the first falsy value in expression and stop evaluating to save the time.
 * In python, 'and' evaluates the operand as truthy or falsy part.
 * But when 'and' encounters the first values as truthy part it return the next operand for evaluation.

```python
a = 1
b = True
# lets print the and operation 
print(a and b) 
print(b and a)
# What will be the output when both are 1 as we know 1 = True and 0 = False in binary
# Output should be same in both the print statements
# But No. 

# (a and b) -> b (True)
# (b and a) -> a (1)
# Because both operands are not empty or false
```

As I said above, How python "and" evaluate operands.

```python
# When encountering falsy return falsy operand
def check():
    print("Function called")
    return True

print(0 and check()) # opt -> 0 
```

*check()* we never execute because and encounted the first operand as falsy part.

```python
def check():
    print("Function called")
    return True

print(check() and 0) # opt -> check() and 0 (as falsy)
```

## 'OR' Operator Short-Circuit

 * In 'or' operator, 'or' evaluates the expression when it encounters the first truthy value in expression and stop evaluating to save the time.
 * When 'or' encounters the first values as truthy part it does not return the next operand for evaluation.

```python
# When encountering falsy return falsy operand
def check():
    print("Function called")
    return True

print(0 or check()) # opt -> check() true
```

```python
# Real Life UseCase
# For Example Assume input_value entered "" this.
# Default value pattern.
input_value = ""
name = input_value and "Guest"
# Here, if input_value is nothing or 0 then name will assign false not "Guest"
name
```

```python
name = input_value or "Guest"
# Here, if input_value is nothing or 0 then name will evaluate false then return to "Guest" for evaluation
name
```

```python
# Tricky Example 
print(([] and "Hello") or "World") # output -> world
# (falsy or "World")
```
