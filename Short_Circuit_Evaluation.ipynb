# Short-Circuit Evaluation

Works on 'and' and 'or' operator

In Python, 'and' and 'or' operator does not evaluate the expression completely.

They stop early when they encounter the decide result.

This is called Short-Circuit Evaluation.

```python
# For example 
print("" and "4" and " ") 
```

## 'AND' Operator Short-Circuit

In and operator, 'and' evaluates the expression when it encounters the first falsy value in expression and stop evaluating to save the time.

In python, 'and' evaluates the operand as truthy or falsy part.

But when 'and' encounters the first values as truthy part it return the next operand for evaluation.

```python
a = 1
b = True
# lets print the and operation 
print(a and b) 
print(b and a)
# What will be the output when both are 1 as we know 1 = True and 0 = False in binary
# Output should be same in both the print statements
# But No. 

# (a and b) -> b (True)
# (b and a) -> a (1)
# Because both operands are not empty or false
```

It must return true or 1 in both the condition *But* AND returned one truthy and one falsy operand. 
    
*       In python, and does not return true or false    
*       It actually returns one of the operand.
*          When encountering 1 in (a and b) it return 1 as true the truthy part so the answer is true
*          When encountering true in (b and a) it return true as the truthy part because non empty or zero value so, answer is 1

```python
# When encountering falsy return falsy operand
def check():
    print("Function called")
    return True

print(0 and check()) # opt -> 0 
```

*check()* we never execute because and encounted the first operand as falsy part.

```python
def check():
    print("Function called")
    return True

print(check() and 0) # opt -> check() and 0 (as falsy)
```

## 'OR' Operator Short-Circuit

In 'or' operator, 'or' evaluates the expression when it encounters the first truthy value in expression and stop evaluating to save the time.

When 'or' encounters the first values as truthy part it does not return the next operand for evaluation.

```python
# When encountering falsy return falsy operand
def check():
    print("Function called")
    return True

print(0 or check()) # opt -> check() true
```

```python
# Real Life UseCase

# Default value pattern
name = input_value ="" and "Guest"
# Here, if input_value is nothing or 0 then name will assign false not "Guest"
name
```

```python
name = input_value ="" or "Guest"
# Here, if input_value is nothing or 0 then name will evaluate false then return to "Guest" for evaluation
name
```

```python
# Tricky Example 
print(([] and "Hello") or "World") # output -> world
# (falsy or "World")
```

```python
# other 
print("1" and 1 and 0) # return 0 
print("1" and 0 and True) # return 0 
print(" " and "Arman" and 534) # return 534 because " " space, "" empty
print("" and "Arman" and 534) # return nothing(space)
```
