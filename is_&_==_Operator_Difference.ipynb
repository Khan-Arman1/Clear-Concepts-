# Problem and Confusion Statement with 'is' and '==' operator

We have to tell the difference



```python
# Suppose mere pass 2 variables hai , Inme mujhe 'is' and '==' operator me difference batana hai 
# Similarity - dono check karte hai value same hai ya nahi fir main difference kya hua inme 

a = 5
b = 5

print(a==b)
print(a is b)
# output true ayega because values same hai
print(type(a) == type(b)) # true
print(type(a) is type(b)) # true
# output esa aya because dono ke data type same hai

# Value change karke dekhte hai
b = '5'
# string liya hai
print()
print(a==b) #false
print(a is b) #false
# But difference kya hai agar dono hi false dikha rahe haitoh means working same hai

# DataType bhi different de diye false dikha raha hai, kyuki agar 'is' datatype bhi check nahi karta toh kya difference hai
print(type(a) == type(b))
print(type(a) is type(b))
```

# Solution - 

### First know the definition of both - 

 ##### **1. '=='** - Check the value, whether they are equal to each other or not.

```python
var1 = "Arman Khan"
var2 = "Khan Arman"
print(var1==var2)
# false bcs == check the characters in string 
# var1[0] != var2[0] -> A != K
# So output will be false
```

```python
# check with same value but differnt types
var1 = 1 # int
var2 = "1" # str
print(var1==var2)
```

In here var1 and var2 are equal in value 
var1 == var2 -> 1 == '1'

But 

They are type difference, Python is a strongly typed language means python consider datatypes as well

So, 1 == '1' but integer 1 != string '1' 

Since, the output is false.

##### **2. 'is' -** checks the variable pointing to the same value. Return true if both variables have the same object.


```python
x = ["apple", "banana"]
y = ["apple", "banana"]
z = x

print(x is z)
print(x is y)
print(x == y)
```

```python
# The result of (x is y) is false because they list is an object and python create a new object for list rathen 
# refrencing the same object
print(id(x))
print(id(y))
print(id(z))
```

```python
# But in integers python refer the same integer to another variable
a = 5
b = 5
c = a

print(a is c)
print(a is b)
print(a == b)

print(id(a))
print(id(b))
print(id(c))
```

```python
# lets see it works with string object or not
a = "Arman Khan"
b = "Arman Khan"
c = a

print(a is c)
print(a is b)
print(a == b)
```

```python
# In Dictionary
dict_1 = {1:'1',2:[1,2]}
dict_2 = {1:'1',2:[1,2]}
dict_3 = dict_1

print(dict_1 is dict_3)
print(dict_1 is dict_2)
print(dict_1 == dict_2)
# If we see the difference
dict_2 = {1:'1',2:[1,'2']}
print(dict_1 == dict_2)
```

##### **Here the main difference in 'is' and '==' operator is - **

1. *'=='* - compares data (values)

2. *'is'* - compares address (id)
