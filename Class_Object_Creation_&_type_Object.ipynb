## Can we manipulate the object creation?

 * But before doing that let us see an example on what we gonna manipulate.

```python
class CustomizedMetaClass(type):
    
    print("Meta class called")

    def __call__(my_class_reference, *args, **kwargs):
        print("__call__() called")
        my_obj = my_class_reference.__new__(my_class_reference,*args,**kwargs)
        
        print("My class MyClass reference, ",my_class_reference)

        if isinstance(my_obj, my_class_reference):
            print(f"Yes, {my_obj} is an instance of {my_class_reference}")
            my_class_reference.__init__(my_obj,*args,**kwargs)
        
        return my_obj        


class MyClass(metaclass=CustomizedMetaClass):
    
    print("MyClass Called ")
    
    def __new__(cls):
        print("__new__() called in MyClass, cls: ",cls)
        return super().__new__(cls)
    
    def __init__(self):
        print("__init__() called, self: ",self)

    
object_variable_1 = MyClass() # This is class call -> type.__call__(cls)
```

## Important know how objects are created?

 * In Python every class creation and initialization is done in 'type metaclass'.
 * When you write -

            class MyClass():
                   pass

   python internally run this-

           MyClass = type('MyClass',(),{})
   
 * **Anatomy of class creation**

      type (
   
   "MyClass", -> Our class name

   (), -> There are actually the base classes that python inherit by default. e.g. - OBJECT class

    {} -> These are the attributes that we use create in our class.

   )

```python
Example = type(
    "A",
    (object,), # takes a tuple argument
    {
        "name":"Arman Khan",
        "age":22,
        "profoession":"Data Science Learner"
    }
)

Example # Example is just a varible that we used to store the reference of that class. 
# Class is actually in the name of 'A'. 
# In Example we just passed that class in memory reference
```

 * When you write -

           object = MyClass()
   Here,

   - object -> Variable that store the reference
   - MyClass() -> Class Call
 * Internally calls type's '__ call __()' method.
 * Evalulation -

        obj = type. __ call __(MyClass)
 * Further -  __ call __(cls)

                |
                | -> __ new __(cls)
                |
                | -> __ init __(self)

 * Afterward call() method call both __ new __() and __ init __().

       MyClass. __ new __(cls)
   cls - class reference
 * Internally this code looks like this -

          obj = cls. __ new __(cls, *args, **kwargs)
 * __ new __() - allocated memory, right.
 * But
 * In reality, this __ new __() never allocates memory. 'Object' class does.
 * __ new __() really use to customize the object creation or to start that process.
 * Then internally

            object. __ new __(cls) will be called.
 * Then,

           cls. __ init __(obj, *args, **kwargs)
 * Now understand that memory is allocated for the object internally but the memory space is empty because all the objects are separate.
 * Our object is not initialized yet. Object doesn't have State and Behaviour.
   
   

```python
class MyMeta(type):

    def __call__(cls,*args,**kwargs):

        my_obj=cls.__new__(cls,*args,**kwargs)
        print(f"MyClass reference cls:",cls)
        # 
        # object class called. object.__new__(cls)
        #
        if isinstance(my_obj,cls):
            print(f"MyClass in memory object reference my_obj",my_obj)
            # manipulate the reference
            my_obj = "Mimiced reference by Arman at <0>"
            cls.__init__(my_obj)
            
        return my_obj

class MyClass(metaclass = MyMeta):

    def __new__(cls):
        print("__new__ called")
        return super().__new__(cls)
    def __init__(self):
        print("__init__ called")
        print(f"Customized object reference self:",self)

obj = MyClass()
obj
```

 * MyClass(metaclass=CustomizedMetaClass) is a way you explicitly define your meta class.

## What happened above? 

 * Actually, our __ new __() -> super(). __ new __() makes a call.
 * There object is being created but we are not passing that object reference to initialize.

## Now lets try we don't let python create object internally.
